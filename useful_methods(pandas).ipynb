```python
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return "generous"
    else:
        return "other"
```

```python
quality(10.34,1.66)
```

```python
df['Tipquality'] = df[['total_bill','tip']].apply (lambda df: quality(df['total_bill'],df['tip']),axis=1)
```

```python
df['Tipquality'] 
```

```python
df['Tipquality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
```
